```python
import os
import math
import json
import matplotlib.pyplot as plt
import numpy as np
import librosa
import math

NoiseCafe=[]
dataset_path='/content/Cafe_-5/drspeech/data/TCDTIMIT/Noisy_TCDTIMIT/Cafe'
TRACK_DURATION = 5 # measured in seconds
SAMPLE_RATE=8000 #Hz
SAMPLES_PER_TRACK = SAMPLE_RATE * TRACK_DURATION
num_mfcc=20
hop_length=1024
n_fft=2048
num_segments=2


samples_per_segment = int(SAMPLES_PER_TRACK / num_segments)
num_mfcc_vectors_per_segment = math.ceil(samples_per_segment / hop_length)

# loop through all genre sub-folder
for dirpath, dirnames, filenames in os.walk(dataset_path):
    print("present dir: ",dirpath)

    # ensure we're processing a genre sub-folder level
    if dirpath is not dataset_path:
        pass
        # process all audio files in genre sub-dir
        for f in filenames:
          if f.endswith(".wav"):
            print("wavfile: ",f)

# load audio file
            file_path = os.path.join(dirpath, f)
            signal, sample_rate = librosa.load(file_path, sr=SAMPLE_RATE)

            # process all segments of audio file
            for d in range(num_segments):

                # calculate start and finish sample for current segment
                start = samples_per_segment * d
                finish = start + samples_per_segment

                # extract mfcc
                mfcc = librosa.feature.mfcc(signal[start:finish], sr=sample_rate, n_mfcc=num_mfcc, n_fft=n_fft, hop_length=hop_length)
                mfcc = mfcc.T
#                 print(num_mfcc_vectors_per_segment)
#                 print(len(mfcc[0]))

                #store only mfcc feature with expected number of vectors
                if len(mfcc) == num_mfcc_vectors_per_segment:
                    NoiseCafe.append(np.log(np.abs(mfcc.tolist())))


```

```python
pickle_train = open("noiseCafe_-5.pickle","rb")
X = pk.load(pickle_train)
pickle_trainlabel = open("CleanCafe_5.pickle","rb")
y = pk.load(pickle_trainlabel)
```

```python
X=np.asarray(X)
y=np.asarray(y)
```

```python
from sklearn.model_selection import train_test_split
X_tra, X_test, y_tra, y_test = train_test_split(X, y, test_size=0.3)
X_train, X_val, y_train, y_val = train_test_split(X_tra, y_tra, test_size=0.3)
```

```python

import keras
from keras import layers
input_img = keras.Input(shape=(X_train.shape[1],X_train.shape[2], 1))
x = layers.Conv2D(64, (3, 3), activation='relu', padding='same',kernel_initializer='glorot_uniform')(input_img)
x = layers.Dropout(0.3)(x)
x = layers.MaxPooling2D((2, 2), padding='same')(x)

x = layers.Conv2D(64, (3, 3), activation='relu', padding='same',kernel_initializer='glorot_uniform')(x)
x = layers.Dropout(0.3)(x)
x= layers.Conv2D(32, (3, 3), activation='relu', padding='same',kernel_initializer='glorot_uniform')(x)
x = layers.Dropout(0.3)(x)
encoded = layers.MaxPooling2D((2, 2), padding='same')(x)


x = layers.Conv2D(32, (3, 3), activation='relu', padding='same',kernel_initializer='glorot_uniform')(encoded)
x = layers.Dropout(0.3)(x)
x = layers.UpSampling2D((2, 2))(x)
x = layers.Conv2D(64, (3, 3), activation='relu', padding='same',kernel_initializer='glorot_uniform')(x)
x = layers.Dropout(0.3)(x)
x = layers.UpSampling2D((2, 2))(x)
decoded = layers.Conv2D(1, (3, 3), activation='relu', padding='same',kernel_initializer='glorot_uniform')(x)

autoencoder = keras.Model(input_img, decoded)
autoencoder.compile(optimizer='adam', loss='mean_squared_error')

```

```python
autoencoder.fit(X_train, y_train,
                epochs=100,
                batch_size=32,
                shuffle=True,
                validation_data=(X_val, y_val))
```

```python
from keras import losses 
from keras import optimizers 
from keras import metrics 

autoencoder.compile(loss = 'mean_squared_error',  
   optimizer = 'adam', metrics = [metrics.categorical_accuracy])
```

```python
# y_train.shape
results = autoencoder.evaluate(X_test, y_test, batch_size=8)
print("test loss, test acc:", results)
```
